```python
documents=[
    "The quick brown fox jumps over the lazy dog.",
    "This is the second line of text.",
    "Another line added for demonstration purposes.",
    "We are adding multiple lines to the list.",
    "This list will contain a few sample sentences.",
    "Just a few more lines to reach the desired count.",
    "This is the seventh line.",
    "And finally, the eighth line."
]
```

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model_name='sentence-transformers/paraphrase-xlm-r-multilingual-v1'
```

```python
model=SentenceTransformer(model_name)
```

```python
document_embeddings=model.encode(documents)
```

```python
len(document_embeddings)
```

```python
for i, embedding in enumerate(document_embeddings):
  print(f"Document {i + 1}")
  print(f"Embedding: {embedding}")
```

```python
query = "What is mentioned in the initial lines of the document?"
```

```python
query_embeddings=model.encode(query)
```

```python
print("Query embedding:", query_embeddings)
```

```python
len(query_embeddings)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities=cosine_similarity(np.array([query_embeddings]),document_embeddings)
```

```python
similarities
```

```python
most_similar_index=np.argmax(similarities)
```

```python
most_similar_index
```

```python
most_similar_document=documents[most_similar_index]
```

```python
most_similar_document
```

```python
query
```

```python
sorted_indices=np.argsort(similarities[0])[::-1]
```

```python
sorted_indices
```

```python
ranked_documents=[(documents[i],similarities[0][i]) for i in sorted_indices]
```

```python
ranked_documents
```

```python
print("Ranked Documents:")
for rank , (document , similarity) in enumerate (ranked_documents , start=1):
  print(f"Rank {rank}:Document - '{document}' , similarity score - {similarity}")
```

```python
print("Top 4 Documents :")
for rank , (document ,similarity ) in enumerate(ranked_documents[:4] , start=1):
  print(f"Rank {rank}:Document - '{document}' , similarity score - {similarity}")
```

```python
!pip install rank_bm25
```

```python
top_4_documents=[doc[0] for doc in ranked_documents[:4]]
```

```python
top_4_documents
```

```python
tokenized_top4_documents=[doc.split() for doc in top_4_documents]
```

```python
tokenized_top4_documents
```

```python
tokenized_query=query.split()
```

```python
tokenized_query
```

```python
from rank_bm25 import BM25Okapi
```

```python
bm25=BM25Okapi(tokenized_top4_documents)
```

```python
bm25_score=bm25.get_scores(tokenized_query)
```

```python
bm25_score
```

```python
sorted_indices2=np.argsort(bm25_score)[::-1]
```

```python
sorted_indices2
```

```python
top_4_documents
```

```python
reranked_documents = [(top_4_documents[i], bm25_score[i]) for i in sorted_indices2]
```
