```python


from youtube_transcript_api import YouTubeTranscriptApi,TranscriptsDisabled
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
```

###STEP1 -INDEXNG


```python

video_id="-BjZmE2gtdo"
try:
  transcript_list=YouTubeTranscriptApi().fetch(video_id,languages=["en"])
  transcript=" ".join(chunk.text for chunk in transcript_list)
  print(transcript)
except TranscriptsDisabled:
  print("transcript disabled")
```

```python
transcript_list
```

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=100,chunk_overlap=40)
chunks=splitter.create_documents([transcript])
```

```python
len(chunks)
```

```python
!pip install -u lan
```

```python
pip install -U langchain-huggingface
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vector_store = FAISS.from_documents(chunks, embeddings)

print("Vector store ready")
```

```python
vector_store.index_to_docstore_id
```

##2-Retrieval

```python
retriever=vector_store.as_retriever(search_type="similarity",search_kwargs={"k":4})
```

```python
retriever
```

```python
retriever.invoke('who is lover')
```

##3-Augmentation

```python
pip install -U transformers accelerate langchain langchain-core langchain-community langchain-huggingface
```

```python
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline

pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    max_new_tokens=200,
    do_sample=True,
    temperature=0.2
)

llm = HuggingFacePipeline(pipeline=pipe)

print(llm.invoke("Explain embeddings in simple words"))
```

```python
prompt=PromptTemplate(
template="""
you are a helpfull assistant.
answer only from the provided transcript context.
if the context is insufficient , just say you dont know.

{context}
Question:{question}
""",
input_variable=['context','question']
)
```

```python
question="is the topic of lover discussed in this video? if yes then where"
retrieved_docs= retriever.invoke(question)
```

```python
context_text="\n\n".join(doc.page_content for doc in retrieved_docs)
```

```python
final_prompt=prompt.invoke({"context":context_text,"question":question})
```

```python
final_prompt
```

##4-Generation

```python
answer=llm.invoke(final_prompt)
print(answer)
```

##Building a Chain

```python
from langchain_core.runnables import RunnableParallel ,RunnablePassthrough ,RunnableLambda
from langchain_core.output_parsers import StrOutputParser
#
```

```python
def format_docs(retrieved_docs):
  context_text="\n\n".join(doc.page_content for doc in retrieved_docs)
  return context_text
```

```python
parllel_chain=RunnableParallel({
    'context':retriever | RunnableLambda(format_docs),
    'question':RunnablePassthrough()
})
```

```python
parllel_chain.invoke("who is lover")
```

```python
parser=StrOutputParser()
```

```python
main_chain=parllel_chain |prompt | llm |parser
```

```python
main_chain.invoke("summarize the video")
```
