```python
from sentence_transformers import SentenceTransformer
embedding_model=SentenceTransformer("thenlper/gte-large")
```

```python


def get_embedding(text:str)->list[float]:
  if not isinstance(text, str) or not text.strip():
    print("attempted to get embedding for empty or non-string text.")
    return []
  embedding=embedding_model.encode(text)
  return embedding.tolist()
```

```python
data["embedding"]=data["fullplot"].apply(get_embedding)
```

```python
def vector_search(user_query,collection):
  query_embedding=get_embedding(user_query)
  print(query_embedding)

  if query_embedding is None:
    return "invalid query"

  pipeline=[
    {
        "$vectorSearch":{
            "index":"vector_index",
            "queryVector": query_embedding,
            "path":"embedding",
            "numCandidates":150,
            "limit":4,
        }
    },
    {
        "$project":{
            "fullplot":1,
            "title":1,
            "genres":1,
            "score":{"$meta":"vectorSearchScore"},
        }
    }
]
  result=collection.aggregate(pipeline)
  return list(result)
```

```python
vector_search("which is the best horror movie ",collection)
```

```python
query="what is the best horror movie ?"
```

```python
def get_search_result(query,collection):
  get_knowledge=vector_search(query,collection)
  search_result=""
  for result in get_knowledge:
    search_result += f"Title:{result.get('title','N/A')},plot:{result.get('fullplot','N/A')}\n"
  return search_result
```

```python
source_information=get_search_result(query,collection)
```

```python
combined_information=f"Query:{query}\nContinous to answer the query by using the search Results :\n{source_information}."
print(combined_information)
```
